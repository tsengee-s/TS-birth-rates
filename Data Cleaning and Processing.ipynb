```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv("C:/Users/sundu/OneDrive - The University of Chicago/Documents/Spring 2023/Time Series/Vital_Stats.csv")
df.head(11)
```

```python
df1 = df.head(10)
df1
```

```python
df2 = df1.drop(columns=['By administrative divisions', 'Item', 'UNIT'])
df2
```

```python
df3 = df2.T
df3 = df3.iloc[:-1]
df3.columns = df3.iloc[0]
df3 = df3[1:]
```

```python
df3
```

```python
df4 = df3.reset_index()
df4.columns = ['Date', 'Live births', 'CBR','Deaths', 'Crude Death Rate', 'Natural increase',
              'Natural increase rate', 'Marriages','Crude Marriage Rate', 'Divorces','Crude Divorce Rate']
df4
```

```python
df4.isna().sum()
```

```python
data1 = df4[['Date', 'Live births', 'Marriages', 'Divorces']].copy()
data2 = df4[['Date', 'CBR', 'Crude Marriage Rate', 'Crude Divorce Rate']].copy()
```

```python
data2 = data2.dropna(subset=['CBR']).reset_index(drop=True)
data2.isna().sum()
```

```python
data1
```

```python
data2
```

```python
data1['Date'] = pd.to_datetime(data1['Date'].str.replace(' Month', ''), format='%Y.%m')
```

```python
data2['Date'] = pd.to_datetime(data2['Date'].str.replace(' Month', ''), format='%Y.%m')
```

```python
data1['Live births'] = data1['Live births'].astype(float)
data1['Marriages'] = data1['Marriages'].astype(float)
data1['Divorces'] = data1['Divorces'].astype(float)
```

```python
data1.dtypes
```

```python
data2['CBR'] = data2['CBR'].astype(float)
data2['Crude Marriage Rate'] = data2['Crude Marriage Rate'].astype(float)
data2['Crude Divorce Rate'] = data2['Crude Divorce Rate'].astype(float)
```

```python
data2.dtypes
```

```python
# Calculate the correlation matrix
correlation_matrix1 = data1.corr()
correlation_matrix2 = data2.corr()
# Print the correlation matrix
print(correlation_matrix1)
print(correlation_matrix2)
```

### EDA

```python
import seaborn as sns
sns.boxplot(data=data1[['Live births', 'Marriages', 'Divorces']])
plt.ylabel('Value')
plt.title('Box Plot of Variables')
plt.show()
```

```python
import seaborn as sns
sns.boxplot(data=data2[['CBR', 'Crude Marriage Rate', 'Crude Divorce Rate']])
plt.ylabel('Value')
plt.title('Box Plot of Variables')
plt.show()
```

```python
# Plot a histogram of a single variable
sns.histplot(data1['Live births'], kde=True)
plt.ylabel('Frequency')
plt.title('Number of live births distribution')
plt.show()
```

```python
sns.histplot(data2['CBR'], kde=True)
plt.ylabel('Frequency')
plt.title('Crude birth rate distribution')
plt.show()
```

Stationarity

```python
from statsmodels.graphics.tsaplots import plot_acf

births_ts = pd.Series(data1['Live births'])
# Plot the ACF
plot_acf(births_ts, lags = 100)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function of Live Births')
plt.show()
```

```python
cbr_ts = pd.Series(data2['CBR'])
# Plot the ACF
plot_acf(cbr_ts, lags = 100)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function Plot of Crude Birth Rate')
plt.show()
```

```python
# Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(data1['Date'], data1['Live births'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series: Live Births')
plt.grid(True)
plt.show()
```

```python
# Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(data2['Date'], data2['CBR'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series: Crude Birth Rate')
plt.grid(True)
plt.show()
```
